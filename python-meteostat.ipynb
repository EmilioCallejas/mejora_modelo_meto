```python
# Data from weather station NOAA

from datetime import datetime
import matplotlib.pyplot as pl
from meteostat import Point, Hourly
from geopy.geocoders import Nominatim
import pandas as pd


#Coordenadas

loc = Nominatim(user_agent="GetLoc") # geololocalizar ciudad de interes
getLoc = loc.geocode("Viña Del Mar / Población Parque Forestal") # Esto lo encontre en la base de datos
print(getLoc.address)

#Lat y long

lat = getLoc.latitude
long = getLoc.longitude
print("Latitude = ", lat)
print("Longitude = ", long)


#para obtener información de estaciones del módulo meteostat
#OBTENER LISTADO DE ESTACIONES

from meteostat import Stations

stations = Stations()
stations = stations.nearby(lat,long)
staion = stations.fetch(40)
print(staion)

#definir periodo de tiempo a utlizar

start = datetime(2016, 1, 1)
end = datetime(2023, 9, 20, 23, 59)


#establecer punto

vina1 = Point(lat,long) #Corresponde a la estacion seleccionada

#obtener datos diarios

data = Hourly(vina1, start, end)
data = data.fetch()

# 
df = pd.DataFrame(data)
columns_names = df.columns.values
print(columns_names)

#mostrar datos
print(data)

# Guarda los datos
df.to_pickle("Vina_Forestal_2016_2023.pkl")

#generar gráfica de temperaturas
data.plot(y=['temp'])
pl.grid()
pl.title("Viña del Mar / Forestal, temperatura del aire")
pl.xlabel("Años")
pl.ylabel("ºC")
pl.show()
```
